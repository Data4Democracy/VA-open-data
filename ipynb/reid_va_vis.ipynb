## Import Data CSV as a Pandas DataFrame

```python
import pandas as pd
import os

clean_data = pd.read_csv(os.path.join('..','data','vets_by_state_clean.csv'))
```

### Show Data

```python
clean_data.head()
```

### Create Scatter Matrix for Relevant Columns

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import matplotlib.pyplot as plt
subset = clean_data.iloc[:,[6,7,8]]
subset.columns = ['Disability\nRate', "Below Poverty Level\nIncome", "Bachelor's\nAttainment"]
pd.plotting.scatter_matrix(
    subset, alpha=0.7, 
    figsize=(9, 9), diagonal='hist'
)
plt.suptitle("Scatter Matrix of Several Important Variables")
plt.show()
```

### Show Boxplot of The same set of variables

```python
with plt.style.context(('ggplot')):
    axis = subset.plot(kind="box")
plt.show()
```
