# Initial VA EDA

```python
#Open dataset information data into a data frame
import os
import pandas as pd

raw_data_frame = pd.read_csv(
    os.path.join(os.path.dirname(os.getcwd()), 
                 'data', 
                 'va_data_inventory_links_checked_20171202.csv')
)
```

```python
#View head of a data frame
from IPython.display import display
display(raw_data_frame.head())
```

```python
#Download an example Excel workbook and load a page into a Pandas dataframe
import urllib
ex_download_url = raw_data_frame.loc[1, 'distribution_downloadURL']
local_excel_workbook = os.path.join(os.path.dirname(os.getcwd()),'data','urban_vets_by_state.xlsx')
urllib.request.urlretrieve(
    ex_download_url,
    local_excel_workbook
)
```

```python
#Read in Excel file and explore Excel Sheets
urban_vets_by_state = pd.read_excel(local_excel_workbook, header=[1, 2])
print(type(urban_vets_by_state))
```
