# Initial VA EDA

```python
#Open dataset information data into a data frame
import os
import pandas as pd

raw_data_frame = pd.read_csv(
    os.path.join(os.path.dirname(os.getcwd()), 
                 'data', 
                 'va_data_inventory_links_checked_20171202.csv')
)
```

```python
#View head of a data frame
from IPython.display import display
display(raw_data_frame.head())
```

```python
#Download an example Excel workbook and load a page into a Pandas dataframe
import urllib
ex_download_url = raw_data_frame.loc[1, 'distribution_downloadURL']
local_excel_workbook = os.path.join(os.path.dirname(os.getcwd()),'data','urban_vets_by_state.xlsx')
urllib.request.urlretrieve(
    ex_download_url,
    local_excel_workbook
)
```

```python
#Read in Excel file and explore Excel Sheets
urban_vets_by_state = pd.read_excel(local_excel_workbook, header=[1, 2], na_values=".")
print(type(urban_vets_by_state))
```

```python
#Visualize and explore vets by state DataFrame
display(urban_vets_by_state)
```

```python
urban_vets_rename = urban_vets_by_state.loc[urban_vets_by_state.index != 'Alabama',:]
new_index = list(urban_vets_rename.index.values[1:])
#Append the new value *in place*
new_index.insert(0, "Alabama")

urban_vets_rename.index = new_index
urban_vets_rename.index.name = 'STATE'
display(urban_vets_rename)
```

```python
urban_vets_states_only = urban_vets_rename.iloc[0:52,:]
```

```python
urban_vets_states_only.rename(columns={'Unnamed: 6_level_1': 'Rural'}, inplace=True)
display(urban_vets_states_only.head())
```

```python
#Stack the table to make columns single index
stacked_table = urban_vets_states_only.stack(level=1)
display(stacked_table.head())
```

```python
clean_data = pd.pivot_table(stacked_table, index='STATE', aggfunc=sum).reset_index()
display(clean_data)
```

```python
#Write out cleaned data as a CSV
clean_data.to_csv(    
    os.path.join(os.path.dirname(os.getcwd()), 
                 'data', 
                 'vets_by_state_clean.csv'),
    index=False
)
```
